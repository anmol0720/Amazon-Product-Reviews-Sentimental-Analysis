```python
from nltk.corpus import stopwords
```

```python
df['reviews.doRecommend']=df['reviews.doRecommend'].replace([True,False],["1","0"])
```

```python
df['reviews.doRecommend']=df['reviews.doRecommend'].astype("int")
```

```python
document=df['reviews.text']
```

```python
stop_words = set(stopwords.words('english')) 
#word_tokens = word_tokenize(document)
filtered_sentence = [w for w in stop_words]   
filtered_sentence = [] 
for w in stop_words: 
        filtered_sentence.append(w)
```

```python
#Special Character Cleaning
sentences=[]
for sentence in document:
    sentences.append(re.sub('[^a-zA-Z0-9\n\.]', ' ', sentence))
```

```python
tfidf= TfidfVectorizer(stop_words= 'english', vocabulary= df['reviews.text'])
X=tfidf.fit_transform(df['reviews.text'])
Y=df['reviews.doRecommend']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model = LogisticRegression()
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=40,test_size=0.8)
```

```python
y_train.dtype
```

```python
l=model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
target_names = ['True','False']
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
